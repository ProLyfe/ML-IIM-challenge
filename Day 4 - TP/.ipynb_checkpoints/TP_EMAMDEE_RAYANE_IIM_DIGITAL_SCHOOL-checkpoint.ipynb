```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.model_selection import learning_curve

# Create a baseline dummy classifier
dummy_model = DummyClassifier(strategy="most_frequent")
dummy_model.fit(X_train, y_train)

# Predict on the test set
y_pred_dummy = dummy_model.predict(X_test)

# Calculate performance metrics
accuracy = accuracy_score(y_test, y_pred_dummy)
recall = recall_score(y_test, y_pred_dummy)
precision = precision_score(y_test, y_pred_dummy)
f1 = f1_score(y_test, y_pred_dummy)

# Print baseline model performance metrics
print("Baseline Dummy Model Performance:")
print("Accuracy:", accuracy)
print("Recall:", recall)
print("Precision:", precision)
print("F1 Score:", f1)

train_sizes, train_scores, val_scores = learning_curve  (
    dummy_model,
    X_train,
    y_train,
    cv=5
)
train_scores_mean = np.mean(train_scores, axis=1)
val_scores_mean = np.mean(val_scores, axis=1)

plt.plot(train_sizes, train_scores_mean, color="b", label="Training Score")
plt.plot(train_sizes, val_scores_mean, color="g", label="Validation Score")

plt.figure(figsize=(10, 6))
plt.title("Learning Curves")
plt.xlabel("Validation")
plt.ylabel("Score")
           
plt.grid()
plt.show()
```
